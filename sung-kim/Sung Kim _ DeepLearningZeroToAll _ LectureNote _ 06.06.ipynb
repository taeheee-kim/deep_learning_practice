# tf2_02_01_linear_regression

**목표 : minimize Cost(W,b)**

Hypothesis Function : H(x) = Wx + b

Cost Function : mean_squared( H(x)-y )

```python
import numpy as np
import tensorflow as tf
```

```python
x_train = [1,2,3,4]
y_train = [0, -1, -2, -3]
```

```python
tf.model = tf.keras.Sequential()
# units == output shape, input_dim == input shape
tf.model.add(tf.keras.layers.Dense(units=1, input_dim=1))
```

```python
# SGD == standard gradient descendent, lr == learning rate
sgd = tf.keras.optimizers.SGD(learning_rate=0.1)
# mse == mean_squared_error, 1/m * sig (y'-y)^2
tf.model.compile(loss='mse', optimizer=sgd)
```

```python
# prints summary of the model to the terminal
tf.model.summary()
```

```python
# fit() executes training
tf.model.fit(x_train, y_train, epochs=200)
```

```python
tf.print(tf.model.weights)
```

W = -0.999008298  /   
b = 0.9970842   /

**model : -0.999008298 x + 0.9970842**    
vs  
**real : - x + 1**  

```python
# predict() returns predicted value
y_predict = tf.model.predict(np.array([5, 4]))
y_test = np.array([[-4],[-3]])
for i in range(2):
    print(y_predict[i],'vs',y_test[i])
```

# tf2_03_01_minimizing_cost_show_graph

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
x_train = [1, 2, 3, 4]
y_train = [0, -1, -2, -3]
```

```python
tf.model = tf.keras.Sequential()
tf.model.add(tf.keras.layers.Dense(units=1, input_dim=1))
```

```python
sgd = tf.keras.optimizers.SGD(learning_rate=0.1)
tf.model.compile(loss='mse', optimizer=sgd)
```

```python
tf.model.summary()
```

```python
# fit() trains the model and returns history of train
history = tf.model.fit(x_train, y_train, epochs=100)
```

```python
y_predict = tf.model.predict(np.array([5, 4]))
print(y_predict)
```

```python
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```
